```python
from textblob import TextBlob
sample = kf.iloc[10].comments
TextBlob(sample).sentiment
```

```python
def sentiment_calc(text):
    try:
        return TextBlob(text).sentiment
    except:
        return None

kf['sentiment'] = kf['comments'].apply(sentiment_calc)
```

```python
#kf[['polarity']] = kf.sentiment.polarity
kf['polarity'] = kf['sentiment'].apply(lambda x: x.polarity)
kf['subjectivity'] = kf['sentiment'].apply(lambda x: x.subjectivity)
kf = kf.drop(['sentiment'],axis=1)
kf.head()
```
